## Figures

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import math
import seaborn as sns

plt.rcParams['font.family'] = 'Avenir Next'
plt.rcParams['font.weight'] = 200 
plt.rcParams['font.size'] = 15
plt.rcParams['axes.titlesize'] = 16 
plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titleweight'] = 200
plt.rcParams['axes.labelweight'] = 200
plt.rcParams['figure.figsize'] = (10, 6) 
curve_color = '#DD4C4F'
plt.rcParams['figure.dpi'] = 300
# curve_color = '#DD4C4F'
```

### Concentration / Anti-concentration of Geometric

```python
p1 = .2
p2 = .8

x = np.arange(50)
def fx(x, p):
    return x * p * (1-p)**(x-1)

width = 0.35
x_pos = np.arange(len(x))

fig, ax = plt.subplots()
ax.bar(
    x_pos - width/2,
    fx(x, p1),
    width=width,
    label=f"p = {p1}",
    alpha=0.8
)

ax.bar(
    x_pos + width/2,
    fx(x, p2),
    width=width,
    label=f"p = {p2}",
    alpha=0.8
)

ax.set_xticks(x_pos[::5])
ax.set_xticklabels(x[::5])
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid(True, alpha = 0.3, linestyle='--')
ax.set_title("Concentration / Anti-Concentration of Geometric r.v.s")
plt.tight_layout()
plt.show()
```

### Concentration / Anti-concetration of Pascal Random Variables

```python
r_arr = np.arange(2, 6)
r_arr = 2 ** r_arr 
p_arr = np.arange(0.1, 0.8, 0.2)

x = np.arange(0, 500)

fig, axes = plt.subplots(4, 1, figsize = (10, 10))
axes = axes.flatten()

colors = ['#DD4C4F', '#4C7CDD', '#4CDD7C', '#DDA84C']

for idx, r in enumerate(r_arr):
    ax = axes[idx]
    
    n_bars = len(p_arr)
    width = 1.0 / (n_bars + 1)
    
    for p_idx, p in enumerate(p_arr):
        x_scipy = x - r
        pmf_vals = scipy.stats.nbinom.pmf(x_scipy, r, p)
        
        positions = x + (p_idx - n_bars/2 + 0.5) * width
        
        ax.bar(
            positions, 
            pmf_vals, 
            width=width, 
            color=colors[p_idx],
            label=f'$p = {p:.1f}$', 
        )
    
    ax.set_title(f'$r = {r}$')
    ax.set_ylabel('$P(T_r = k)$')
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.set_xlim(0, 300) 

axes[0].legend(loc='upper right', fontsize=9)
axes[-1].set_xlabel('$k$')
plt.suptitle('Pascal Concentration / Anti-Concentration', 
             fontsize=15, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()
```
